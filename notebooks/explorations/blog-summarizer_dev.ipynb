## Step 4: Chunking — Handling Long Documents Properly

Naively truncating a post from the beginning produces poor summaries because 
blog posts don't follow the "inverted pyramid" structure that BART was trained on.

The solution is **chunking**:
1. Split the full post into overlapping sections
2. Summarize each section independently  
3. Combine those summaries into a final summary

This is called map-reduce summarization and is the standard approach for 
documents that exceed a model's token limit.

```python
def chunk_text(text, chunk_size=1000, overlap=100):
    """
    Split text into overlapping chunks.
    
    Args:
        text: The full text to chunk
        chunk_size: Characters per chunk
        overlap: Characters of overlap between chunks
                 (prevents losing context at boundaries)
    
    Returns:
        List of text chunks
    """
    chunks = []
    start = 0
    
    while start < len(text):
        end = start + chunk_size
        chunk = text[start:end]
        chunks.append(chunk)
        start = end - overlap  # step back by overlap amount
    
    return chunks

# Split our post into chunks
chunks = chunk_text(sample_post, chunk_size=1000, overlap=100)

print(f"Post split into {len(chunks)} chunks")
for i, chunk in enumerate(chunks):
    print(f"Chunk {i+1}: {len(chunk)} characters")
```

## Step 5: Summarize Each Chunk

Now we run each chunk through BART independently, then combine those 
individual summaries into a single final summary. This ensures every 
part of the post is represented.

```python
def summarize_chunks(chunks, summarizer):
    """
    Summarize each chunk independently.
    
    Args:
        chunks: List of text chunks
        summarizer: The Hugging Face pipeline
    
    Returns:
        List of summaries, one per chunk
    """
    chunk_summaries = []
    
    for i, chunk in enumerate(chunks):
        print(f"Summarizing chunk {i+1} of {len(chunks)}...")
        
        # Skip chunks that are too short to summarize meaningfully
        if len(chunk.split()) < 30:
            print(f"  Chunk {i+1} too short, skipping")
            continue
            
        summary = summarizer(
            chunk,
            max_length=60,
            min_length=20,
            do_sample=False
        )
        
        chunk_summaries.append(summary[0]['summary_text'])
        print(f"  Done: {summary[0]['summary_text'][:80]}...")
    
    return chunk_summaries

chunk_summaries = summarize_chunks(chunks, summarizer)

print(f"\n✓ Generated {len(chunk_summaries)} chunk summaries")
```

## Step 6: Combine Chunk Summaries into a Final Summary

With each chunk summarized individually, we now concatenate those summaries 
and run them through BART one final time to produce a single coherent summary 
of the entire post.

This two-pass approach (summarize chunks → summarize the summaries) is the 
core of map-reduce summarization.

```python
def final_summary(chunk_summaries, summarizer):
    """
    Combine chunk summaries into one final summary.
    
    Args:
        chunk_summaries: List of individual chunk summaries
        summarizer: The Hugging Face pipeline
    
    Returns:
        Final combined summary string
    """
    # Join all chunk summaries into one block of text
    combined = " ".join(chunk_summaries)
    
    print("Chunk summaries combined:")
    print(f"Combined length: {len(combined.split())} words")
    print(f"\n--- COMBINED CHUNK SUMMARIES ---\n")
    print(textwrap.fill(combined, width=80))
    
    print("\n\nGenerating final summary...")
    
    summary = summarizer(
        combined,
        max_length=150,
        min_length=50,
        do_sample=False
    )
    
    return summary[0]['summary_text']

result = final_summary(chunk_summaries, summarizer)

print("\n\n=== FINAL SUMMARY ===\n")
print(textwrap.fill(result, width=80))
```

```python
def chunk_by_paragraph(text, max_chunk_words=150):
    """
    Split text by paragraphs rather than character count.
    Respects the natural structure of the content.
    
    Args:
        text: Full post text
        max_chunk_words: Max words before starting a new chunk
    
    Returns:
        List of chunks, each containing complete paragraphs
    """
    # Split on double newlines (paragraph breaks)
    paragraphs = [p.strip() for p in text.split('\n\n') if p.strip()]
    
    chunks = []
    current_chunk = []
    current_word_count = 0
    
    for paragraph in paragraphs:
        paragraph_words = len(paragraph.split())
        
        # If adding this paragraph exceeds limit, save current chunk and start new one
        if current_word_count + paragraph_words > max_chunk_words and current_chunk:
            chunks.append('\n\n'.join(current_chunk))
            current_chunk = [paragraph]
            current_word_count = paragraph_words
        else:
            current_chunk.append(paragraph)
            current_word_count += paragraph_words
    
    # Don't forget the last chunk
    if current_chunk:
        chunks.append('\n\n'.join(current_chunk))
    
    return chunks

# Rechunk by paragraph
para_chunks = chunk_by_paragraph(sample_post, max_chunk_words=150)

print(f"Post split into {len(para_chunks)} paragraph-based chunks")
for i, chunk in enumerate(para_chunks):
    print(f"Chunk {i+1}: {len(chunk.split())} words")
```

```python
# Summarize the paragraph-based chunks
para_chunk_summaries = summarize_chunks(para_chunks, summarizer)

print(f"\n✓ Generated {len(para_chunk_summaries)} chunk summaries")
```

```python
# Generate the final summary from paragraph-based chunks
para_result = final_summary(para_chunk_summaries, summarizer)

print("\n\n=== FINAL SUMMARY (paragraph-based chunking) ===\n")
print(textwrap.fill(para_result, width=80))
```

## Findings & Analysis

### What Worked
- Paragraph-based chunking preserved complete thoughts and produced coherent 
  individual chunk summaries
- The combined chunk summaries accurately represent the full post content
- The model correctly identified the key thesis: proposing collective nouns 
  for engineering disciplines

### What Didn't Work
- The second-pass compression (summarizing the summaries) loses too much 
  context and nuance
- BART was fine-tuned on CNN/DailyMail news articles which follow a strict 
  inverted-pyramid structure — creative, humorous, narrative writing is 
  outside its training distribution
- The humor and voice of the original post is completely lost

### Key Insight
For this type of content, the **combined chunk summaries may be more useful 
than a single final summary**. Rather than forcing one final compression pass,
presenting section-by-section summaries preserves more meaning.

## Step 7: Final Summarizer — Section-by-Section Output

Based on our analysis, forcing a second compression pass loses too much context.
The most useful output for a narrative blog post is a structured 
section-by-section summary that preserves the flow of the original content.

We'll wrap everything into a single clean function that represents our finished 
summarizer.

```python
def summarize_blog_post(text, summarizer, max_chunk_words=150):
    """
    Summarize a blog post using paragraph-based chunking.
    
    Rather than forcing a second compression pass (which loses context),
    this function returns structured section summaries that preserve the
    flow and meaning of the original post.
    
    Args:
        text: Full blog post text
        summarizer: Hugging Face summarization pipeline
        max_chunk_words: Max words per chunk (default 150)
    
    Returns:
        dict containing:
            - chunk_summaries: List of section summaries
            - stats: Basic statistics about the post
    """
    # Basic stats
    stats = {
        'word_count': len(text.split()),
        'char_count': len(text),
        'chunk_count': 0
    }
    
    # Chunk by paragraph
    chunks = chunk_by_paragraph(text, max_chunk_words)
    stats['chunk_count'] = len(chunks)
    
    # Summarize each chunk
    print(f"Summarizing {len(chunks)} sections...")
    summaries = []
    
    for i, chunk in enumerate(chunks):
        print(f"  Section {i+1} of {len(chunks)}...")
        
        if len(chunk.split()) < 30:
            continue
            
        summary = summarizer(
            chunk,
            max_length=60,
            min_length=20,
            do_sample=False
        )
        
        summaries.append(summary[0]['summary_text'])
    
    return {
        'chunk_summaries': summaries,
        'stats': stats
    }


# Run it
result = summarize_blog_post(sample_post, summarizer)

# Display results cleanly
print("\n")
print("=" * 60)
print("BLOG POST SUMMARY REPORT")
print("=" * 60)
print(f"Word count:    {result['stats']['word_count']}")
print(f"Sections:      {result['stats']['chunk_count']}")
print("=" * 60)

for i, summary in enumerate(result['chunk_summaries']):
    print(f"\nSection {i+1}:")
    print(textwrap.fill(summary, width=60, initial_indent="  "))

print("\n" + "=" * 60)
```

## Conclusions & Next Steps

### What We Built
A blog post summarizer using Facebook's BART model with paragraph-based 
chunking that produces structured section-by-section summaries.

### Key Learnings
- Naive truncation produces misleading summaries for long-form content
- Character-based chunking breaks mid-sentence and confuses the model
- Paragraph-based chunking respects natural content boundaries and produces
  significantly better results
- For creative/narrative writing, section-by-section summaries outperform
  forced single-paragraph compression
- BART performs best on factual, news-style content — creative and humorous 
  writing is outside its training distribution

### Limitations
- BART was fine-tuned on CNN/DailyMail news articles — it loses the voice 
  and humor of creative writing
- Section summaries are useful but not a true abstractive summary of the 
  whole post
- Model inference is slow on CPU (~5-7 seconds per chunk)

### Potential Next Steps
- Test alternative models (PEGASUS, T5) on the same content for comparison
- Fine-tune BART on a custom dataset of blog posts for better style matching
- Build a simple Gradio UI to make the summarizer usable without the notebook

## Appendix: Loading Posts From Files

To summarize a new blog post, save it as a `.txt` file in the `notebooks/` 
folder and run the cell below.

```python
def load_post_from_file(filename):
    """
    Load a blog post from a text file in the notebooks directory.
    
    Args:
        filename: Name of the .txt file (e.g. 'my-post.txt')
    
    Returns:
        Post text as a string
    """
    filepath = os.path.join(os.getcwd(), filename)
    
    if not os.path.exists(filepath):
        raise FileNotFoundError(f"Could not find {filepath}")
    
    with open(filepath, 'r', encoding='utf-8') as f:
        text = f.read()
    
    print(f"Loaded: {filename}")
    print(f"Word count: {len(text.split())}")
    
    return text

# Example usage — uncomment and swap in your filename:
# post = load_post_from_file('my-post.txt')
# result = summarize_blog_post(post, summarizer)
```
